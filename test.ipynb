# 循环语句

```python
# continue : 当条件成立，退出当前一次循环，继而执行下一次循环
# 吃5个苹果 -- 循环； 吃到第3个吃出一个虫子，第三个不吃了，没吃饱，继续吃4和5个苹果 -- 只有第三个不吃
i = 1
while i <= 5:
    # 条件
    if i == 3:
        print('吃出一个大虫子，这个苹果不吃了')
        # 如果使用continue，在continue之前一定要修改计数器，否则进入死循环
        i += 1
        continue
        #跳转到while开头判断
    print(f'吃了第{i}个苹果')
    i += 1

i = 1
while i <= 5:
    # 条件
    if i == 3:
        print('吃出一个大虫子，这个苹果不吃了')
        # 如果使用continue，在continue之前一定要修改计数器，否则进入死循环
        i += 1
        break#直接跳出循环
    print(f'吃了第{i}个苹果')
    i += 1
```

# 循环嵌套

```python
#自拟三角形星号
i=1#控制行

while i<=5:
    j=1#控制列
    while j<=i :
        print('*',end=' ')
        j+=1
    print(' ')
    i+=1
#九九乘法表
i=1
while i<=9 :
    j=1
    while j<=i :
        print(f'{j}*{i}={i*j}',end=' ')
        j+=1
    print('\n')#注意语句的位置不要出错
    i+=1
```

## 3.6局部和全局调用域

全局作用域代码不能使用局部变量
局部作用域代码可以访问全局变量
一个函数的局部作用域代码不能使用其他局部作用域变量
不同作用域函数可以使用相同名称命名不同的变量

### 3.6.1局部变量不能在全局作用域使用

```python
def spam():
    eggs = 25
spam()
# print(eggs) #报错：NameError: name 'eggs' is not defined
```

### 3.6.2局部作用域不能使用其他局部作用域内的变量

```python
def fx():
    x=120
    fy()
    print(x)

def fy():
    z=102
    x=20

fx()#仅能调用fx自己变量中的x，不能调用fy中的x
```

### 3.6.3全局变量可以再局部作用域读取

```python
def fx():
    print(y)
y=120
fx()
print(y)
#注意，y=120此处为全局变量，无论是fx()中print(y)指令还是最后的print（y）指令都可以调用全局定义好的y=20
```

### 3.6.4名称相同的局部变量和全局变量

```python
def fx():
    m = '迪迦'
    print(m)

def fy():
    m = '泰罗'
    print(m)
    fx()
    print(m)

m = '赛文'
fy()
print(m)
#运行函数，最先使用局部变量
```

## 3.7global语句

```python
#使用global定义全局变量
def fx():
    global m
    m = '赛罗' #全局变量

def fy():
    m = '迪迦' #局部变量

def fz():
    print(m) #全局变量

m = '贝利亚' #全局变量
print(m)
fx()
print(m)
#调用fx后，m为全局变量，值为'赛罗'
```
